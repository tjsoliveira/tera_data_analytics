```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
d1 = pd.read_csv('dataset3.csv', sep=',', encoding='windows-1252')
d2 = pd.read_csv('dataset2.csv', sep=',', encoding='windows-1252')
```

```python
# a. Quantas linhas e quantas colunas esse dataset possui?
d1
```

```python
d2
```

```python
df = d1.merge(d2, how='inner', on='Id')
```

```python
df = df[['Id', 'Name_x', 'Platform_x', 'Year_of_Release_x', 'Genre_x',
       'Publisher_x', 'Critic_Score', 'Critic_Count', 'User_Score',
       'User_Count', 'Developer', 'Rating', 'NA_Sales', 'EU_Sales',
       'JP_Sales', 'Other_Sales', 'Global_Sales']]
df.columns
```

```python
df = df.rename(columns={
    'Name_x': 'Name', 
    'Platform_x': 'Platform',
    'Year_of_Release_x': 'Year_of_Release',
    'Genre_x': 'Genre',
    'Publisher_x': 'Publisher'}, errors="raise")
df
```

```python
corr = df.corr()
corr
```

```python
plt.figure(figsize=(30, 10))
sns.heatmap(corr)
```

Pelo heatmap apresentado, percebe-se a forte influência das vendas nos Estados Unidos nas vendas no restante do mundo.
Após os Estados Unidos, a Europa é quem mais influencia nas vendas dos jogos pelo mundo
